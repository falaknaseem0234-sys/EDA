Question 2: Explain overfitting and underfitting in ML. How can you detect and prevent
them?
Hint: Discuss bias-variance tradeoff, cross-validation, and regularization techniques.
Ans 2:1. Introduction

In Machine Learning (ML), the goal is to build a model that generalizes well — meaning it performs accurately not just on the training data, but also on unseen test data.
Two common problems that affect model performance are overfitting and underfitting.

2. Overfitting

Definition:
Overfitting occurs when a model learns the training data too well, capturing not only the underlying patterns but also noise and random fluctuations.
As a result, it performs very well on the training set but poorly on new, unseen data.

Example:
A decision tree that grows too deep may memorize the training samples but fail to predict new cases correctly.

Symptoms:

Very low training error, but high test error.

Model is too complex (many parameters).

Graphically:
Training accuracy ↑
Testing accuracy ↓

Causes:

Too complex model (deep neural networks, large trees).

Too few training samples.

Too many irrelevant features.

3. Underfitting

Definition:
Underfitting occurs when a model is too simple to capture the underlying patterns in the data.
It performs poorly on both the training and test datasets.

Example:
Using a simple linear model to fit non-linear data.

Symptoms:

High training error and high test error.

Model fails to capture relationships in the data.

Causes:

Model too simple (few parameters).

Insufficient training.

Inappropriate features.

4. Bias-Variance Tradeoff

The bias-variance tradeoff explains the balance between underfitting and overfitting:

Term	Meaning	Effect
Bias	Error from overly simplistic assumptions in the model	Leads to underfitting
Variance	Error from model’s sensitivity to small fluctuations in training data	Leads to overfitting

Goal: Find a balance between bias and variance for optimal generalization.

Ideal Model: Low bias + low variance → performs well on both training and test data.

5. Detection of Overfitting and Underfitting
Method	Overfitting Detection	Underfitting Detection
Training vs Test Accuracy	High training accuracy, low test accuracy	Both training and test accuracy are low
Learning Curves	Training loss ↓, validation loss ↑	Both training and validation losses remain high
Cross-Validation	High variance across folds	Consistently poor performance across folds
6. Prevention Techniques
(a) Cross-Validation

Use k-fold cross-validation to estimate model performance more reliably.

Helps identify overfitting early by checking consistency across data subsets.

(b) Regularization

Adds a penalty to large model coefficients to control complexity.

L1 Regularization (Lasso): Encourages sparsity (feature selection).

L2 Regularization (Ridge): Penalizes large weights evenly.

Elastic Net: Combination of L1 and L2.

(c) Early Stopping

In iterative models (e.g., neural networks), stop training when validation error starts increasing even if training error decreases.

(d) Pruning (for Decision Trees)

Remove unnecessary branches that add complexity and noise.

(e) Dropout (for Neural Networks)

Randomly deactivate some neurons during training to prevent co-adaptation.

(f) Feature Selection

Remove irrelevant or redundant features to simplify the model.

(g) Increase Training Data

More diverse data helps the model learn general patterns rather than memorizing.

(h) Simplify or Tune the Model

For overfitting: reduce model complexity.

For underfitting: increase model complexity or add more features.

7. Summary Table
Aspect	Overfitting	Underfitting
Model Complexity	Too high	Too low
Training Error	Low	High
Test Error	High	High
Bias	Low	High
Variance	High	Low
Solution	Regularization, cross-validation, pruning	Add features, increase complexity, train longer
8. Conclusion

Overfitting and underfitting are two opposite problems in ML that affect model generalization.

Overfitting → Model learns noise (high variance).

Underfitting → Model misses patterns (high bias).
By applying cross-validation, regularization, early stopping, and careful model selection, we can achieve a good bias-variance balance and improve model performance on unseen data.

Question 3:How would you handle missing values in a dataset? Explain at least three
methods with examples.
Hint: Consider deletion, mean/median imputation, and predictive modeling.
Ans 3: Introduction

In real-world datasets, missing values are very common due to errors in data collection, transmission, or human entry.
Handling missing data properly is essential because it affects the accuracy, bias, and reliability of Machine Learning models.

Missing values can occur in three ways:

MCAR (Missing Completely at Random): Missingness has no relation to any variable.
Example: A sensor randomly fails to record temperature.

MAR (Missing at Random): Missingness depends on other observed variables.
Example: Older patients less likely to report income.

MNAR (Missing Not at Random): Missingness depends on the missing value itself.
Example: People with very high income choose not to disclose it.

2. Methods to Handle Missing Values

Below are three major methods (with examples) commonly used:

(A) Deletion Methods
1. Listwise Deletion (Complete Case Analysis)

Definition: Remove all rows that contain one or more missing values.

Use When: The dataset is large and missing values are rare (<5%).

Example:

Name	Age	Salary
A	25	30,000
B	30	NaN
C	28	40,000

→ Remove row B (since Salary is missing).

Advantages:

Simple and easy to implement.

Maintains consistency in analysis.

Disadvantages:

Can lead to loss of information.

May introduce bias if data is not MCAR.

2. Pairwise Deletion

Definition: Use available data for each analysis without removing entire rows.

Example: When calculating correlation between “Age” and “Salary,” use only records with both values available.

Use When: Only certain analyses require complete data and others can use partial information.

(B) Imputation Methods
1. Mean/Median/Mode Imputation

Definition: Replace missing values with a statistical measure of that column (mean, median, or mode).

Use When: Missing values are small and data is numeric or categorical.

Example:

Name	Age
A	25
B	NaN
C	27

→ Mean Age = (25 + 27)/2 = 26 → Replace NaN with 26.

Advantages:

Simple and fast.

Preserves dataset size.

Disadvantages:

Can reduce variance in data.

May distort relationships between variables.

Choice:

Use mean for normal data.

Use median for skewed data.

Use mode for categorical data.

2. K-Nearest Neighbors (KNN) Imputation

Definition: Missing values are filled based on similar samples (neighbors) in the dataset.

Example: If a student’s marks are missing, take the average of marks from the 5 most similar students.

Advantages: More accurate than simple mean/median.

Disadvantages: Computationally expensive for large datasets.

(C) Predictive Modeling Methods
1. Regression Imputation

Definition: Predict missing values using a regression model built from other available features.

Example:
Suppose Salary is missing. We can predict it using a regression model:
Salary = f(Age, Education, Experience)

Advantages:

Uses relationships between variables.

More accurate than simple statistical methods.

Disadvantages:

Adds complexity.

Can overfit if missingness is high.

2. Multiple Imputation

Definition: Generates multiple possible imputed datasets, analyzes each, and combines results for more reliable estimates.

Advantages: Preserves data variability.

Disadvantages: Computationally intensive.

3. Choosing the Right Method
Condition	Recommended Method
Few missing values	Deletion (Listwise/Pairwise)
Numeric continuous data	Mean/Median imputation
Categorical data	Mode imputation
Complex datasets	Regression or KNN imputation
High missing ratio	Predictive modeling or Multiple imputation
4. Example Summary
Method	Description	Example	Best Use Case
Deletion	Remove missing rows	Drop rows with NaN values	Small % missing
Mean/Median Imputation	Replace missing with average	Replace missing age with 26	Numerical data
Predictive Modeling	Use regression/KNN to estimate	Predict missing salary	Large datasets with patterns
5. Conclusion

Handling missing data is a critical preprocessing step in data analysis.

Deletion is simple but risks losing data.

Mean/Median Imputation is easy but can reduce data variability.

Predictive Modeling offers better accuracy but adds complexity.

The best approach depends on the amount of missing data, data type, and modeling goals.
Effective handling ensures better accuracy, generalization, and reliability of machine learning models.

Question 4:What is an imbalanced dataset? Describe two techniques to handle it
(theoretical + practical).
Hint: Discuss SMOTE, Random Under/Oversampling, and class weights in models.

Ans 4:1. Introduction

In classification problems, an imbalanced dataset occurs when the number of observations in one class is much higher or lower than in other classes.
This causes the model to become biased toward the majority class, leading to poor performance on the minority (rare) class, which is often the most important one (e.g., fraud detection, disease prediction).

2. Example of an Imbalanced Dataset
Transaction ID	Is Fraud?
1	No
2	No
3	No
4	Yes

→ Out of 4 samples, 3 are “No” and only 1 is “Yes.”
So, the dataset is imbalanced (75% vs 25%).

If a model predicts “No” for all cases, it would still be 75% accurate — but completely useless for detecting frauds.

3. Why Imbalance is a Problem

The model becomes biased toward the majority class.

Accuracy becomes misleading (high accuracy but poor recall for minority class).

Important minority patterns get ignored.

Hence, handling imbalance is essential for fair and accurate prediction.

4. Techniques to Handle Imbalanced Datasets

Here are two major techniques (both theoretical and practical):

(A) Random Sampling Methods
(i) Random Oversampling

Theory:

Increase the number of minority class samples by duplicating existing ones.

Balances the dataset by having equal class proportions.

Example:
If you have 90 “No” and 10 “Yes” samples → duplicate “Yes” samples until both reach 90.

Advantages:

Simple and effective.

Retains all majority class data.

Disadvantages:

May cause overfitting (duplicates may reduce model generalization).

Practical Implementation (Python):

from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=42)
X_res, y_res = ros.fit_resample(X, y)

(ii) Random Undersampling

Theory:

Reduce the majority class by randomly removing samples until both classes are balanced.

Example:
If you have 900 “No” and 100 “Yes” → randomly remove 800 “No” records → 100 “No” and 100 “Yes”.

Advantages:

Reduces training time.

Simple to apply.

Disadvantages:

May lose valuable information from the majority class.

Practical Implementation (Python):

from imblearn.under_sampling import RandomUnderSampler
rus = RandomUnderSampler(random_state=42)
X_res, y_res = rus.fit_resample(X, y)

(B) Synthetic Data Generation – SMOTE (Synthetic Minority Oversampling Technique)
Theory:

SMOTE creates synthetic (artificial) samples of the minority class rather than duplicating existing ones.

It selects a minority sample and one of its k-nearest neighbors and generates new points between them.

Example:
If two minority class samples have feature values 10 and 14, SMOTE might create a synthetic value 12 (interpolated point).

Advantages:

Avoids overfitting caused by simple duplication.

Produces more general and realistic synthetic samples.

Disadvantages:

May create noisy data if minority class is highly variable.

Slightly increases computation time.

Practical Implementation (Python):

from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(X, y)

(C) Using Class Weights in Models
Theory:

Assign higher penalty (weight) to misclassifying minority class samples during training.

The model then gives more importance to rare cases.

Formula:
Weight = (Total samples) / (Number of classes × Samples in that class)

Example:
In a binary classification with 900 “No” and 100 “Yes”:
Weight for “Yes” = 1000 / (2 × 100) = 5 → “Yes” gets 5× more importance.

Advantages:

No data duplication or deletion.

Simple to use in algorithms like Logistic Regression, SVM, and Random Forest.

Practical Implementation (Python):

from sklearn.linear_model import LogisticRegression
model = LogisticRegression(class_weight='balanced')
model.fit(X_train, y_train)

5. Evaluation Metrics for Imbalanced Data

Since accuracy can be misleading, use:

Precision = TP / (TP + FP)

Recall (Sensitivity) = TP / (TP + FN)

F1-Score = 2 × (Precision × Recall) / (Precision + Recall)

ROC-AUC Curve – Area under the Receiver Operating Characteristic curve.

These metrics give a clearer picture of how well the model handles the minority class.

6. Summary Table
Method	Type	Advantages	Disadvantages	Best Use Case
Random Oversampling	Data-level	Simple, retains data	Risk of overfitting	Small datasets
Random Undersampling	Data-level	Fast, easy	Data loss	Large datasets
SMOTE	Data-level	Realistic synthetic samples	Can add noise	Numeric, moderate imbalance
Class Weights	Algorithm-level	No data modification	Needs model support	Logistic Regression, SVM
7. Conclusion

An imbalanced dataset can severely reduce model performance on important minority classes.
To overcome this:

Data-level techniques like Oversampling, Undersampling, and SMOTE balance the data.

Algorithm-level techniques like Class Weight Adjustment make the model fairer.

Combining these approaches with proper evaluation metrics ensures a balanced, fair, and accurate model.

Question 5: Why is feature scaling important in ML? Compare Min-Max scaling and

Hint: Explain impact on distance-based algorithms (e.g., KNN, SVM) and gradient
Ans 5:1. Introduction

In Machine Learning, feature scaling (also called normalization) is the process of bringing all features to a similar range or scale.
This is crucial because many algorithms rely on distances or gradients, which are sensitive to the magnitude of features.

2. Why Feature Scaling is Important

When features have different units or ranges (e.g., age in years vs. income in rupees), large-valued features dominate model learning, causing bias and poor convergence.

Example:
Feature	Range
Age	18 – 60
Income	₹10,000 – ₹10,00,000

If not scaled, algorithms may treat “Income” as more important simply because of its larger numerical range.

3. Algorithms Affected by Feature Scaling
(A) Distance-Based Algorithms

K-Nearest Neighbors (KNN): Uses Euclidean distance.
Features with larger values dominate the distance metric.

Support Vector Machine (SVM): Sensitive to feature magnitudes in computing the optimal hyperplane.

K-Means Clustering: Cluster centers can be distorted by unscaled features.

(B) Gradient-Based Algorithms

Linear/Logistic Regression, Neural Networks: Feature scaling helps in faster convergence of gradient descent.
Without scaling, gradients oscillate and take longer to reach minima.

(C) PCA (Principal Component Analysis)

PCA depends on variance. If features aren’t scaled, large-valued features dominate the principal components.

Question 6:  Compare Label Encoding and One-Hot Encoding. When would you prefer
one over the other?
Hint: Consider categorical variables with ordinal vs. nominal relationships.
Ans 6:1. Introduction

In Machine Learning, most algorithms cannot process categorical data directly.
Hence, we must convert categorical (text-based) values into numerical format so that algorithms can interpret them.

Two common encoding techniques used for this are:

Label Encoding

One-Hot Encoding

Each has its advantages and is suitable for different types of categorical data — ordinal or nominal.

2. Types of Categorical Variables
Type	Description	Example
Ordinal	Categories have a natural order or ranking	Low < Medium < High
Nominal	Categories have no order; all are unique	Red, Blue, Green

The choice of encoding depends on which type of variable we have.

3. Label Encoding
Definition:

Label Encoding converts each category into a unique integer value.

Example:

Size	Encoded
Small	0
Medium	1
Large	2

Here, categories are replaced with numerical labels.

When to Use:

Suitable for ordinal categorical data (ordered categories).
Example: education levels, product ratings, customer satisfaction.

Advantages:

Simple and efficient.

Preserves ordinal relationship (order of values).

Requires less memory (single column only).

Disadvantages:

Creates a false sense of order for nominal data.
Example: Encoding Red=0, Blue=1, Green=2 may imply Green > Red, which is meaningless.

Can cause models (like Linear Regression or KNN) to misinterpret relationships.

Python Example:
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X['Size'] = le.fit_transform(X['Size'])

4. One-Hot Encoding
Definition:

One-Hot Encoding converts categorical values into binary vectors (0s and 1s).
Each category gets its own new column.

Example:

Color	Red	Blue	Green
Red	1	0	0
Blue	0	1	0
Green	0	0	1
When to Use:

Best for nominal categorical data (no order among categories).
Example: gender, color, city, brand name.

Advantages:

Avoids false ordering.

Works well with distance and tree-based algorithms.

Disadvantages:

Increases dataset dimensionality (especially with many categories).

May lead to “curse of dimensionality” and higher computation time.

Python Example:
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse_output=False)
X_encoded = ohe.fit_transform(X[['Color']])

5. Comparison Table
Aspect	Label Encoding	One-Hot Encoding
Definition	Assigns each category a unique integer	Creates binary columns for each category
Output	Single column of integers	Multiple binary columns
Best For	Ordinal data (ordered categories)	Nominal data (unordered categories)
Order Preservation	Preserves order (if any)	Does not preserve order
Dimensionality	Low (1 column)	High (many columns)
Interpretation	Risk of false order for nominal data	No risk of false order
Example	Size → {Small:0, Medium:1, Large:2}	Color → Red:[1,0,0], Blue:[0,1,0], Green:[0,0,1]
When to Use	Ordinal features	Nominal features
Disadvantage	Misleads model for nominal data	Expands dataset size
6. When to Prefer Each Method
Situation	Preferred Encoding	Reason
Data has natural order (e.g., “Low”, “Medium”, “High”)	Label Encoding	Keeps order relationship intact
Data has no order (e.g., “Red”, “Blue”, “Green”)	One-Hot Encoding	Avoids implying false ranking
Using Tree-based models (Decision Tree, Random Forest)	Either	Trees handle label-encoded data well
Using Distance/Linear models (KNN, SVM, Linear Regression)	One-Hot Encoding	Avoids bias due to numeric ordering
7. Example Scenario

Dataset:

High School	30,000
Graduate	50,000

➡ Education is ordinal → use Label Encoding.

Dataset:

Delhi	200

Kolkata	180

➡ City is nominal → use One-Hot Encoding.

8. Conclusion

Both Label Encoding and One-Hot Encoding are essential preprocessing techniques for handling categorical variables.

Label Encoding → for ordinal data (order matters).

One-Hot Encoding → for nominal data (no order).

Choosing the correct method ensures that the model correctly interprets relationships between categories and improves prediction performance.

Question 7:  Google Play Store Dataset
a). Analyze the relationship between app categories and ratings. Which categories have the
highest/lowest average ratings, and what could be the possible reasons?
Dataset: https://github.com/MasteriNeuron/datasets.git
(Include your Python code and output in the code box below.)
Ans 7:Goal

To find which app categories have the highest and lowest average ratings and understand the possible reasons behind these differences.

Algorithm (Step-by-Step)

Import Libraries: Use pandas for data handling, matplotlib and seaborn for visualization.

Load Dataset: Read googleplaystore.csv into a pandas DataFrame.

Data Cleaning:

Remove duplicate rows.

Remove rows with missing Category or Rating.

Convert Rating to numeric.

Keep only ratings between 0 and 5.

Compute Average Rating: Group data by Category and calculate the mean rating.

Sort Results: Sort the categories by average rating in descending order.

Visualization: Plot a bar chart of average ratings by category.

Interpretation: Analyze which categories have high/low ratings and why.
Bar chart:
Shows average ratings per category, with Education highest and Entertainment/Finance lowest.

Interpretation and Insights
Observation	Possible Reason
High-rated categories (Education, Books, Art & Design, Health)	Apps provide value, learning, or creativity; reliable and well-structured; users are satisfied.
Low-rated categories (Finance, Dating, Tools, Entertainment)	Apps may have bugs, privacy issues, poor UI, or excessive ads; Entertainment ratings vary due to personal preferences.
Conclusion

Average app ratings vary significantly by category.

Educational and informative apps get the highest ratings due to usefulness and stability.

Finance, Tools, and Dating apps get lower ratings, likely due to performance, privacy, or user experience issues.

Developers should focus on usability, reliability, and content quality to improve ratings.

```python
!git clone https://github.com/MasteriNeuron/datasets.git
```

```python
# Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load Dataset
# Updated path to the dataset after cloning the repository
df = pd.read_csv("datasets/googleplaystore.csv")

# Step 3: Data Cleaning
df = df.drop_duplicates()                     # Remove duplicates
df = df.dropna(subset=['Rating', 'Category']) # Drop missing values
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')
df = df[(df['Rating'] >= 0) & (df['Rating'] <= 5)] # Keep valid ratings

# Step 4: Compute Average Rating by Category
category_ratings = df.groupby('Category')['Rating'].mean().sort_values(ascending=False)

# Step 5: Display Top and Bottom Categories
print("Top 5 Categories by Average Rating:\n")
print(category_ratings.head(5))
print("\nLowest 5 Categories by Average Rating:\n")
print(category_ratings.tail(5))

# Step 6: Visualization
plt.figure(figsize=(12,6))
sns.barplot(x=category_ratings.values, y=category_ratings.index, palette='viridis')
plt.title("Average App Rating by Category", fontsize=14)
plt.xlabel("Average Rating")
plt.ylabel("Category")
plt.show()
```

Question 9: Flight Price Prediction Dataset
a) How do flight prices vary with the days left until departure? Identify any exponential price

(Include your Python code and output in the code box below.)
Ans 9:Compare flight prices across airlines for a route (e.g., Delhi → Mumbai) to identify cheap vs premium airlines.

Algorithm (Theoretical Steps)
Step	Description
Step 1: Filter dataset for specific route (e.g., Source=Delhi & Destination=Mumbai).
Step 2: Group by Airline and calculate average price.
Step 3: Sort airlines by average price to identify cheap and premium carriers.

Step 5: Explain why some airlines are cheaper or premium (services, brand, timing, etc.).Cheaper Airlines: IndiGo, GoAir, SpiceJet → low-cost carriers with fewer amenities.

Insights:

Conclusion

Days Left Analysis: Prices increase sharply near departure; 30–60 days in advance is optimal for booking.

Airline Comparison: Low-cost carriers consistently offer cheaper prices, while full-service airlines are premium due to better amenities.

Strategy: Book early and choose airline based on budget vs comfort requirements.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf

# Load your dataset
df = pd.read_csv("datasets/flight_price.csv") # Corrected file path

# Quickly inspect
print("Columns:", df.columns)
print(df.head())
print(df.info())

# (a) Price vs DaysLeft

# Clean a bit: drop rows with missing Price or DaysLeft
df_clean = df.dropna(subset=['days_left', 'price']) # Corrected column names to lowercase

# Aggregate average & median
days_price = df_clean.groupby('days_left')['price'].agg(['mean', 'median', 'count']).reset_index() # Corrected column names to lowercase

plt.figure(figsize=(10,5))
sns.lineplot(data=days_price, x='days_left', y='mean', label='Mean Price') # Corrected column names to lowercase
sns.lineplot(data=days_price, x='days_left', y='median', label='Median Price') # Corrected column names to lowercase
plt.title("Average / Median Price vs Days Left")
plt.xlabel("Days Left Until Departure")
plt.ylabel("Price")
plt.grid(True)
plt.legend()
plt.show()

# Fit an exponential‑decay style curve: Price ~ a * exp(-b * DaysLeft) + c
def exp_decay(x, a, b, c):
    return a * np.exp(-b * x) + c

# Use only days with decent counts
mask = days_price['count'] > 20
xdata = days_price.loc[mask, 'days_left'].values # Corrected column name to lowercase
ydata = days_price.loc[mask, 'mean'].values # Corrected column name to lowercase

from scipy.optimize import curve_fit

try:
    popt, pcov = curve_fit(exp_decay, xdata, ydata, p0=(1000, 0.01, 100))
    print("Fitted parameters:", popt)
    days_price['predicted'] = exp_decay(days_price['days_left'], *popt) # Corrected column name to lowercase

    # Compute percentage change in predicted
    days_price['pct_change'] = days_price['predicted'].pct_change().fillna(0)
    print("Days with biggest jumps:\n", days_price.sort_values('pct_change', ascending=False).head(10))

    # Plot observed vs predicted
    plt.figure(figsize=(8,4))
    plt.scatter(xdata, ydata, label='Observed')
    xx = np.linspace(min(xdata), max(xdata), 100)
    yy = exp_decay(xx, *popt)
    plt.plot(xx, yy, 'r-', label='Fitted Curve')
    plt.xlabel('DaysLeft')
    plt.ylabel('Price')
    plt.legend()
    plt.title("Observed vs Fitted Exponential Decay Curve")
    plt.show()
except Exception as e:
    print("Curve fitting failed:", e)

# Based on this, you can see where the slope (or pct_change) becomes steep – that region is “price surge” zone.

# (b) Compare airlines for a route (Delhi → Mumbai example)

# Adjust names to match your dataset
src = 'Delhi'
dst = 'Mumbai'

# Filter route
route = df_clean[(df_clean['source_city'] == src) & (df_clean['destination_city'] == dst)] # Corrected column names to lowercase
print("Route data size:", route.shape)
print(route['airline'].unique()) # Corrected column name to lowercase

if route.empty:
    print("No data for this route. Check your city names / spelling.")
else:
    # Summary stats
    airline_stats = route.groupby('airline')['price'].agg(['count', 'mean', 'median', 'min', 'max']).reset_index() # Corrected column names to lowercase
    print("Airline stats sorted by mean price:")
    print(airline_stats.sort_values('mean'))

    # Boxplot
    plt.figure(figsize=(12,6))
    sns.boxplot(data=route, x='airline', y='price') # Corrected column names to lowercase
    plt.title(f"Price Distribution by Airline: {src} → {dst}")
    plt.xticks(rotation=45)
    plt.xlabel("Airline")
    plt.ylabel("Price")
    plt.show()

    # Compare for specific DaysLeft slices
    for d in [30, 20, 10, 5]:
        subset = route[route['days_left'] == d] # Corrected column name to lowercase
        if subset.empty:
            print(f"DaysLeft = {d}: no data")
        else:
            print(f"DaysLeft = {d}:")
            print(subset.groupby('airline')['price'].agg(['mean', 'median']).sort_values('mean')) # Corrected column names to lowercase
            print()

    # Regression: Price ~ Airline + DaysLeft + interaction
    try:
        mdl = smf.ols("price ~ C(airline) + days_left + C(airline):days_left", data=route).fit() # Corrected column names to lowercase
        print(mdl.summary())
    except Exception as e:
        print("Regression failed:", e)
```
